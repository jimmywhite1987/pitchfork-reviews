```python
import pandas as pd
```

```python
red_wine = pd.read_csv("Data/red_wine_0to400_anyrating.csv")
```

```python
sparkling_wine = pd.read_csv("Data/sparkling_wine_0to400_anyrating.csv")
```

```python
white_wine = pd.read_csv("Data/white_wine_0to400_anyrating.csv")
```

```python
red_wine
```

```python
red_wine.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
red_wine['Price'] = red_wine['Price'].str.replace('$','')
```

```python
red_wine['NumberOfRatings'] = red_wine['NumberOfRatings'].str.replace('ratings','')
```

```python
red_wine
```

```python
red_wine['Vintage'] = red_wine['Name'].str.split().str[-1]
```

```python
red_wine['Name'] = red_wine.Name.str.rsplit(' ',1).str[0]
```

```python
red_wine
```

```python
red_wine = red_wine.dropna(how = 'all')
```

```python
red_wine
```

```python
red_wine.drop(red_wine[red_wine.Price.str.contains(r'[ratings]')].index, inplace = True)
```

```python
red_wine.to_csv('Data/clean_red.csv', index=False)
```

```python
white_wine
```

```python
white_wine.drop(['Unnamed: 0'], axis = 1, inplace = True)
white_wine['Price'] = white_wine['Price'].str.replace('$','')
white_wine['NumberOfRatings'] = white_wine['NumberOfRatings'].str.replace('ratings','')
white_wine['Vintage'] = white_wine['Name'].str.split().str[-1]
white_wine['Name'] = white_wine.Name.str.rsplit(' ',1).str[0]
white_wine = white_wine.dropna(how = 'all')
white_wine.drop(white_wine[white_wine.Price.str.contains(r'[ratings]')].index, inplace = True)
```

```python
white_wine
```

```python
sparkling_wine.drop(['Unnamed: 0'], axis = 1, inplace = True)
sparkling_wine['Price'] = sparkling_wine['Price'].str.replace('$','')
sparkling_wine['NumberOfRatings'] = sparkling_wine['NumberOfRatings'].str.replace('ratings','')
sparkling_wine['Vintage'] = sparkling_wine['Name'].str.split().str[-1]
sparkling_wine['Name'] = sparkling_wine.Name.str.rsplit(' ',1).str[0]
sparkling_wine = sparkling_wine.dropna(how = 'all')
sparkling_wine.drop(sparkling_wine[sparkling_wine.Price.str.contains(r'[ratings]')].index, inplace = True)
```

```python
sparkling_wine
```

```python
white_wine.to_csv('Data/clean_white.csv', index=False)
sparkling_wine.to_csv('Data/clean_sparkling.csv', index=False)
```
